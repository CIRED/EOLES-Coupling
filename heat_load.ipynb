```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
heat_load_profile = pd.read_csv("eoles/inputs/hourly_profiles/heat_load_profile.csv")
```

```python
heat_load_profile.columns
```

```python
heat_load_profile["residential_space_weekday_percentage_BDEW"].tolist()
```

```python
fig = plt.figure()
ax = plt.subplot(111)
sns.lineplot(heat_load_profile, x="hour", y="residential_space_weekday_percentage_BDEW", label="weekday BDEW")
sns.lineplot(heat_load_profile, x="hour", y="residential_space_weekend_percentage_BDEW", label="weekend BDEW")
sns.lineplot(heat_load_profile, x="hour", y="residential_space_percentage_valentin", label="profile valentin")

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_ylabel("Heat load profile (%)")
```
